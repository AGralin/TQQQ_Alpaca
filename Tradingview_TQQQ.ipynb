```python
import pandas as pd
from pathlib import Path
import hvplot.pandas
```

```python
# read csv into a dataframe
TQQQ_df = pd.read_csv(
    Path("TQQQ_3.csv"), 
    parse_dates = True, 
    infer_datetime_format = True
)
# Convert Unix timestamp to datetime
TQQQ_df['time'] = pd.to_datetime(TQQQ_df['time'], unit='s')
TQQQ_df.set_index('time', inplace=True)
TQQQ_df.head()
```

```python
TQQQ_df.count()
```

```python
# Plotting the closing prices
plot = TQQQ_df['close'].hvplot.line(
    title='TQQQ Closing Prices (5 second)',
    xlabel='Date',
    ylabel='Price',
    width=800,
    height=500 
)
plot
```

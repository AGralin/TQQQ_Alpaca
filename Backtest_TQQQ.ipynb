```python
import os
from datetime import datetime, timedelta
from dotenv import load_dotenv
import pandas as pd
import hvplot.pandas
import requests
import alpaca_trade_api as tradeapi

%matplotlib inline
```

```python
# Load Alpaca API keys from .env file
load_dotenv()
API_KEY = os.environ.get('ALPACA_API_KEY')
API_SECRET = os.environ.get('ALPACA_SECRET_KEY')
BASE_URL = 'https://paper-api.alpaca.markets'  # Change this if you're not using paper trading

api = tradeapi.REST(API_KEY, API_SECRET, base_url=BASE_URL)

# Calculate the date range
## end_date = datetime.now()  # Today's date
## start_date = end_date - timedelta(days=7)  # 7 days ago
end_date = "2023-10-27"
start_date = "2023-10-24"

# Ensure the dates are in 'America/New_York' timezone
start_date = pd.Timestamp(start_date, tz="America/New_York").isoformat()
end_date = pd.Timestamp(end_date, tz="America/New_York").isoformat()

# Fetch the data
bars = api.get_bars(
    'TQQQ', 
    tradeapi.rest.TimeFrame.Minute, 
    start_date, 
    end_date
).df

# View the data
bars.head(15)
```

```python
# Plotting the closing prices
plot = bars['close'].hvplot.line(
    title='TQQQ Closing Prices (1 Minute)',
    xlabel='Date',
    ylabel='Price',
    width=800,
    height=500 
)
plot
```
